```python
from tensorflow.keras.models import load_model 
model = load_model('asdfinal.model')
#model = load_model('asd20epochs-85.33%.model')
#model.summary()
```

```python
import numpy as np
import matplotlib.pyplot as pt
# import cv2
from tensorflow.keras.preprocessing import image
test_image = image.load_img("C:\\Users\\Darshan Gowda C\\Desktop\\Major project\\ASD project codes\\Project\\test\\autistic\\010.jpg", target_size = (224,224))
pt.imshow(test_image)
test_image = image.img_to_array(test_image)
test_image=test_image/255
test_image=np.expand_dims(test_image, axis = 0)
result = model.predict(test_image)
print(result)
if result[0][0]>result[0][1]:
    print("The image is classified as Autistic with accuracy of:",round(((result[0][0])*100),2),'percent')
else:
    print("The image is classified as Non-Autistic with accuracy of:",round(((result[0][1])*100),2),'percent')
```
